```python
csv_path = '/mnt/video_data/pexels/metadata/results_400k_train_rfcap_pexels_reformat_clean.csv'
```

```python
with open(csv_path, 'r') as f:
    lines = f.readlines()
```

```python
# FPS distribution
fps = [line.split(',')[-4] for line in lines[1:]]
fps = [float(f) for f in fps]
```

```python
# Resolution distribution (height)
res_height = [line.split(',')[-2] for line in lines[1:]]
res_height = [int(h) for h in res_height]
```

```python
# Resolution distribution (width)
res_width = [line.split(',')[-1].strip() for line in lines[1:]]
res_width = [int(w) for w in res_width]
```

```python
# Aspect ratio distribution (width/height)
res_ratio = [w/h for w, h in zip(res_width, res_height)]
```

```python
vid_lens = []
for idx, line_ in enumerate(lines):
    # print(line_)
    if idx == 0:
        continue
    line = line_.strip()
    fps = float(line.split(',')[-4])
    num_frames = float(line.split(',')[-3])
    vid_len = num_frames / fps # in seconds
    vid_lens.append(vid_len)
    
```

```python
# plot histogram of video lengths with vid_lens
import matplotlib.pyplot as plt
import numpy as np
plt.hist(vid_lens, bins=100)
plt.xlabel('Video length (s)')
plt.ylabel('Frequency')
# limit x range from 0 to 200
plt.xlim(0, 200)
plt.title('Histogram of video lengths')
# save to pdf with dpi 300
plt.savefig('hist_vid_lens.pdf', dpi=300)
plt.show()
```

```python
# plot histogram of fps with fps
plt.hist(fps, bins=100)
plt.xlabel('FPS')
plt.ylabel('Frequency')
plt.title('Histogram of FPS')
# save to pdf with dpi 300
plt.savefig('hist_fps.pdf', dpi=300)
plt.show()
```

```python
# plot histogram of resolution height with res_height
plt.hist(res_height, bins=100)
plt.xlabel('Resolution height')
plt.ylabel('Frequency')
plt.title('Histogram of Resolution Height')
# save to pdf with dpi 300
plt.savefig('hist_res_height.pdf', dpi=300)
plt.show()
```

```python
# plot histogram of resolution width with res_width
plt.hist(res_width, bins=100)
plt.xlabel('Resolution width')
plt.ylabel('Frequency')
plt.title('Histogram of Resolution Width')
# save to pdf with dpi 300
plt.savefig('hist_res_width.pdf', dpi=300)
plt.show()
```

```python
# plot histogram of resolution ratio with res_ratio
plt.hist(res_ratio, bins=100)
plt.xlabel('Resolution ratio (width/height)')
plt.ylabel('Frequency')
plt.title('Histogram of Resolution Ratio')
# save to pdf with dpi 300
plt.savefig('hist_res_ratio.pdf', dpi=300)
plt.show()
```
